```python
# استيراد المكتبات
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# قراءة ملف البيانات
data = pd.read_csv(r"C:\Users\Administrator\Desktop\student_dataset_clean.csv")
```

```python
# المتوسط الحسابي
print(np.mean(data["Final"]))
```

```python
# الوسيط
print(np.median(data["Final"]))
```

```python
# الانحراف المعياري
print(np.std(data["Final"]))
```

```python
# اصغر قيمة
print(np.min(data["Final"]))
```

```python
# اكبر قيمة
print(np.max(data["Final"]))
```

```python
# Bar Chart
plt.bar(data["Major"], data["Final"])
plt.title("Final Markes per Major")
plt.xlabel("Major")
plt.ylabel("final grade")
plt.show()
```

```python
# Line Chart
plt.plot(data["Final"])
plt.title("Final Markes Trend")
plt.xlabel("student index")
plt.ylabel("markes")
plt.show()
```

```python
# Histogram
plt.hist(data["Final"])
plt.title("Distribution of final markes")
plt.xlabel("grade range")
plt.ylabel("frequency")
plt.show()
```
